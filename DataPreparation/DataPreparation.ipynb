# Testing preprocess_image Function
This function takes an image as input and applies several preprocessing steps to it, including filtering, grayscale conversion, thresholding, binary conversion, and color inversion if necessary. The output is a preprocessed image.

```python
from DataPreparation import preprocess_image

# Load an image
img = ski.io.imread('../Data/fonts-dataset/Lemonada/299.jpeg')

# Preprocess the image
preprocessed_img = preprocess_image(img)

# Display the preprocessed image
ski.io.imshow(preprocessed_img)
```

# Testing preprocess_and_save_images Function
This function preprocesses a batch of images starting from a given index. It applies the `preprocess_image(img)` function to each image in the batch, and then saves the preprocessed images to a specified directory. The batch size can be set by the user.

```python
from DataPreparation import preprocess_and_save_images

# Preprocess and save 25 images
preprocess_and_save_images(start_index = 0, batch_size = 10)
```
