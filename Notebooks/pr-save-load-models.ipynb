```python
from joblib import dump
from joblib import load
```

```python
def save_Kmeans_model(kmeans):
    # Save the KMeans model
    dump(kmeans, 'kmeans_model_2.joblib')
```

```python
def save_SVM_model(svm):
    # Save the SVM model
    dump(svm, 'svm_model_2.joblib')
```

```python
def load_Kmeans_model():
    # Load the KMeans model
    return load('kmeans_model_2.joblib')
```

```python
def load_SVM_model():
    # Load the SVM model
    return load('svm_model_2.joblib')
```
